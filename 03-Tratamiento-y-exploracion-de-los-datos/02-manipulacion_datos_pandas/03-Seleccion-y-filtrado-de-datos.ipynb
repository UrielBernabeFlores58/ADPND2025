### Seleccion y filtrado de datos

El filtrado permite seleccionar solo las filas que cumplen una o varias condiciones 

```python
import pandas as pd
df = pd.read_csv('./data/linkedin-jobs-usa-modificado.csv')
```

```python
df['onsite_remote'] == 'remote'
```

```python
type(df['onsite_remote'] == 'remote')
```

```python
df_remote = df[df['onsite_remote'] == 'remote']
```

```python
df_remote['onsite_remote']
```

```python
df_remote[['title', 'company', 'onsite_remote']]
```

```python
columnas_visualizar = ['title', 'company', 'onsite_remote']
df_remote[columnas_visualizar]
```
