## Subconjunto de columnas 

Hay muchas formas de subconjuntar un Dataframe, quizas la mas facil es utilizar operadores relacionales para devolver true o false para pasarlos entre corchetes .

`dogs[dogs['height_cm'] > 60]`
`dogs[dogs['color'] == "tan"]`

Puedes filtrar por varias condiciones utilizando el operador & 

`dogs[(dogs['height_cm'] > 60) & (dogs['color'] == 'tan')]`

**Instrucciones**

1. Filtrar homelessness para los casos en los que el numero de individuals sea superior a diez mil, asignalos a `ind_gt_10k`. Visualizar el resultado

2. Filtra homelessnes para los casos en los que la region es Mountain, asignandola a mountain_reg. Visualiza el resultado

3. Filtra homelessness para los casos en los que el numero de family_members sea inferior a mil y la region sea `Pacifico`, asignandolo a fam_lt_1k_pac. Visualizar el resultado

```python
# Filtrar homelessness para los casos en los que el numero de individuals sea superior a diez mil, asignalos a ind_gt_10k. Visualizar el resultado
ind_gt_10k = homelessness[homelessness["individuals"] > 10000]
print(ind_gt_10k.head(10))  # Display the first few rows of the filtered Data
```

```python
# Filtra homelessness para los casos en los que el numero de family_members sea inferior a mil y la region sea `Pacifico`, asignandolo a fam_lt_1k_pac. Visualizar el resultado
fam_lt_1k_pac = homelessness[(homelessness["family_members"] < 1000) & (homelessness["region"] == "Pacific")]
print(fam_lt_1k_pac.head(10))  # Display the first few
```

## A単adir nuevas columnas

No te quedes solo con los datos que te dan, en su lugar, se pueden a単adir nuevas colmnas aun DataFrame. A esto se le llama 'transformacion', 'mutacion' o 'Ingeneria de Caracteristicas'. 

Se pueden crear colmnas nuevas desde cero, pero es habitual derivarias de otras columnas, por ejemplo, sumando columnas o cambiando sus unidades.

**Instrucciones**

- A単ade una nueva columna a homelessness, denominada `Total`, que contenga la suma de la columnas individuals y family_members.

- A単ade otra columna a homelessness, dominada p_homeless, que contenga la proporcion de la poblacion de personas sin hogar total con respecto a poblacion total de cada estado stage_pop.

```python
# Agregar una columna total como la suma de individuals y family_members
homelessness['total'] = homelessness['individuals'] + homelessness['family_members']

# Agregar una columna p_homeless como proporcion del total de homeless
# de la poblacion para el estado de la poblacion (state_pop)

homelessness['p_homeless'] = homelessness['total'] + homelessness['state_pop']

# Visualizar el resultado
homelessness.head(10)

```

Hasta ahora se han visto los 4 tipos mas comunes de manipulacion de datos: `Ordenar filas`, `subconjuntar columnas`, `subconjuntar filas`, y `Anadir nuevas columnas`.

En este ejercicio, responderas la pregunta: `"Que estado tiene el mayor numero de personas sin hogar por cada 10 000 habitantes?"`, combina las nuevas habilidades en pandas para averiguarlo.

**Instrucciones**

- Anade una columa a homelessness, indiv_per_10k, que contenga el numero de personas sin hogar por cada dies mul de personas en cada estado, utilizando state_pop para la poblacion del estado.

- Subconjunta de filas en las que indiv_per_10k sea superiro a 20, asignandolas a high_homelessness.

- Ordena high_homelessness en orden descendente invid_per_10k, asignando a high_srt, Seleccionando solo las columnas state e indiv_per_10k de high_homelessness_srt y guardalas como result y Visualiza result.

```python
# Crear la columna indiv_per_10k
homelessness['indiv_per_10k'] = 10000 * homelessness['individuals'] / homelessness['state_pop']

# subconjuntar filas para indiv_per_10k mayores que 20

high_homelessness = homelessness[homelessness['indiv_per_10k']>20]

# Ordenar high_homelessness por indiv_per_10k

result = high_homelessness.sort_values('indiv_per_10k', ascending=False)

# Visualizar el Resultado
result
```
