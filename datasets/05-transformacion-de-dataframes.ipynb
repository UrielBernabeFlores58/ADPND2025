## Añadir Nuevas Columas

No te quedes solo con los datos que te dan en su lugar, se pueden añadir nuevas columnas a un DataFrame
A esto se le llama `Transformacion`, `Mutacion` e `Ingeneria de Caracteristicas`


se pueden crear nuevas columnas desde cero, pero es habitual derivarlas de otras columnas
por ejemplo, sumando columnas o cambiando sus unidades.

**Instrucciones**

- _Añade una nueva columna a homeless, denominada `total`, que contenga la suma de las columnas individuals y family members_

- _Añade otra columna a homelessness, denomidada p-homeless, que contenga la proporcion de la poblacion de personas sin hogar, con respecto a poblacion total de cada estado stage-pop_

```python
#Agregar una columna total como la suma de individuals y family members
homelessness['total'] = homelessness['individuals'] + homelessness['family_members']

# Agregar columna p_homelessness como proporcion del total de homelessness de la poblacion para el state de la poblacion
homelessness['p_homeless'] = homelessness['total'] / homelessness['state_pop']

#vizualizar el resultado
homelessness[['region', 'state', 'state_pop', 'p_homeless', 'total']].head(10)
```
